Queue dengan List

```python
# Python program to demonstrare queue implementation using list
queue = []

# append() function to enqueue in the queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('initial queue')
print(queue)

# pop(0) function to dequeue element from in FIFO order
print('\nElements dequeued from queue')
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('\nQueue after removing elements')
print(queue)
```

Queue dengan collections.deque

```python
# Python program to demonstrare queue implementation using collections.deque
from collections import deque

queue = deque()

# append() function to enqueue element in the queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('initial queue:')
print(queue)

# popleft() function to dequeue element from in FIFO order
print('\nElements dequeued from queue')
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print('\nQueue after removing elements')
print(queue)
```

Queue dengan queue.Queue

```python
# Python program to demonstrare queue implementation using queue module
from queue import Queue

# initializing a queue
qu = Queue(maxsize=5)

# qsize() show the number of element in the queue
print(qu.qsize())

# put() function to enqueue element in the queue
qu.put('a')
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o')

print("Full: ", qu.full())
print("Size: ", qu.qsize())

# get() function to dequeue element from queue
print('\nElements dequeued from queue')
print(qu.get())
print(qu.get())
print(qu.get())

print("\nEmpty: ", qu.empty())
print("Full: ", qu.full())
```

Queue dengan single Linked List

```python
class Queue:
    def __init__(self):
        self.queue = list()

    # insert method to add element
    def addtoqu(self, dataval):
        if dataval not in self.queue:
            self.queue.insert(0, dataval)
            return True
        return False

    # pop method to remove element
    def removefromq(self):
        if len(self.queue) > 0:
            return self.queue.pop()
        return "No elements in Queue!"

    def size(self):
        return len(self.queue)


TheQueue = Queue()
print(TheQueue.addtoqu("Jan"))
print(TheQueue.addtoqu("Feb"))
print(TheQueue.addtoqu("March"))
print(TheQueue.addtoqu("April"))
print(TheQueue.size())
print(TheQueue.removefromq())
print(TheQueue.removefromq())
print(TheQueue.size())
```
